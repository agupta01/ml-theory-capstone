```python
import hashlib
import os
```

```python
def data_loader(url,download=True) -> bool:
    
    if "data/" not in os.listdir():
        os.makedirs("data")
        
    fname = hashlib.sha256(url.encode()).hexdigest()
    
    if download and os.path.join("data/",fname) in os.listdir("data/"):
        print(f"Dataset already downloaded: data/{fname}")
        return False
    
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser') 
    else:
        print("Failed to retrieve data from website")
    
    res = [i.text.strip().replace("\n","").replace("\r","") for i in soup.find_all('p')]
    res = "\n".join(res)

    with open(fname, "w") as file:
        file.write(res)
    
    print("Dataset successfully downloaded")
    return True
    
    
```
